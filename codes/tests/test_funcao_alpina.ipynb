```python
def uniform_crossover(A, B, P):
    for i in range(len(P)):
        if P[i] < 0.5:
            A[i], B[i] = B[i], A[i]
    return A, B
```

```python
A = np.array([8, 9])
B = np.array([1, 5])
P = np.random.rand(2)
print(P)
```

```python
uniform_crossover(A, B, P)
```

```python
def single_p_crossover(A, B, X):
    A_new = np.append(A[:X], B[X:])
    B_new = np.append(B[:X], A[X:])
    return A_new, B_new
```

```python
X = 1
single_p_crossover(A, B, X)
```
