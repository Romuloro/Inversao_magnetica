```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(-0.1,0.1,100)


def ricker(t, f):
    return (1 - 2*(np.pi*f*t)**2) * np.exp(-(np.pi*f*t)**2)


fx = ricker(t, 5)
```

```python
fig = plt.figure(figsize=(8, 8))
plt.plot(t, fx)
#plt.ylim(0, 11)
plt.title('5 Hz', fontsize=20)
plt.xlabel('$Time(s)$', fontsize=10)
plt.grid()
plt.show()
```
