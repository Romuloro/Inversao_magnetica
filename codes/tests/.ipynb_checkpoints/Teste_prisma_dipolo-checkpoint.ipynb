```python
import numpy as np
import matplotlib.pyplot as plt
import sys
a = sys.path.append('../modules/')
```

```python
import auxiliars, prism, sphere
```

## Etapa 1: Definicão das coordenadas de Observação:

```python
nx = 100 # numero de pontos em x e ....
ny = 100 # ... em y
shape = (nx,ny) 
x = np.linspace(-15000.0, +15000.0, nx) # coordenada x em metros
y = np.linspace(-10000.0, +10000.0, ny) # coordenada y em metros
z = -200.0 # coordenada z em metros

# Ploting all results 
plt.close('all')
plt.figure(figsize=(20,12))
X,Y = np.meshgrid(x,y) # grid de observacoes
plt.plot(X,Y,'.r')
plt.show()
```

## Etapa 2: Simulação do campo Principal na região das observações:

```python
I = 50.03    # inclinacao do campo principal em graus
D = -7.70      # declinacao do campo principal em graus
Fi = 46800.0 # Intensidade do campo principal (nT)

# Campo principal variando com as posicao F(X,Y):
F = Fi + 0.05*X + 0.08*Y # nT

# Calculo das componentes cartesianas do versor F:

# conversao de graus para radiandos:
incl = np.deg2rad(I)
decl = np.deg2rad(D)
#-----------------------------------
Fx = np.cos(incl)*np.cos(decl)
Fy = np.cos(incl)*np.sin(decl)
Fz = np.sin(incl)
```

## Etapa 3: Definição das propriedades das fontes crustais (dipolo e prisma):

```python
# localizacao no espaco (sistema cartesiano) (dipolo):
xc = 5000.0
yc = 5000.0
zc = 1000.0
Mi  = 10.0 # intensidade da magnetizacao em A/m
raio = 325.0 # m

inc = 80.0 # inclinacao do momento de dipolo
dec = 70.0 # declinacao do momento de dipolo

# criacao de uma lista com todas as propriedades do dipolo:
dipolo = [xc, yc, zc, raio]
print (dipolo)
```

```python
# coordenadas dos vertices (corners) do prisma, em metros:
x1,x2  = (-10000.0, -8000.0) 
y1,y2  = (-5000.0, -3000.0)
z1,z2  = (1000.0,2010.0)   # z eh positivo para baixo!

inc2 = -30.0 # magnetizacao puramente induzida
dec2 = 35.0 
Mi2  = 10.0 # intensidade da magnetizacao em A/m

xs = [x1, x1, x2, x2, x1]
ys = [y1, y2, y2, y1, y1]
fonte_crustal_mag = [x1,x2,y1,y2,z1,z2,Mi]
```

```python
esfera = plt.Circle((xc, yc), raio, color ='blue', label='contornos do dipolo', ls= 'solid')

fig, ax = plt.subplots(figsize=(20,12))
#plt.figure(figsize=(20,12))
ax.add_artist(esfera)
plt.plot(X,Y, '.r')
plt.plot(xs,ys,'k-', label='Contornos do prisma')
plt.legend()
#plt.savefig('destribuicao.png', format='png')
plt.show()
```

# Calculo da Anomalia de Campo Total exata:

```python
#Calculo do Campo total DIPOLO
bx = sphere.sphere_bx(X, Y, z, dipolo, Mi, inc, dec)
by = sphere.sphere_by(X, Y, z, dipolo, Mi, inc, dec)
bz = sphere.sphere_bz(X, Y, z, dipolo, Mi, inc, dec)
```

```python
#Calculo do Campo total PRISMA
bx2 = prism.prism_bx(X, Y, z, fonte_crustal_mag, I,D, inc2, dec2) 
by2 = prism.prism_by(X, Y, z, fonte_crustal_mag, I,D, inc2, dec2)  
bz2 = prism.prism_bz(X, Y, z, fonte_crustal_mag, I,D, inc2, dec2)
```

```python
CT =  np.sqrt((F*Fx + bx)**2 + (F*Fy + by)**2 + (F*Fz + bz)**2)
ACT = CT - F


CT2 =  np.sqrt((F*Fx + bx2)**2 + (F*Fy + by2)**2 + (F*Fz + bz2)**2)
ACT2 = CT2 - F
```

```python
# graficos
plt.close('all')
plt.figure(figsize=(17,6))
#******************************************************
plt.subplot(1,2,1)
plt.contourf(Y, X, CT, 20, cmap = plt.cm.RdBu_r)
plt.title('Campo Total (nT)', fontsize = 12)
plt.xlabel('East (m)', fontsize = 10)
plt.ylabel('North (m)', fontsize = 10)
plt.colorbar()
#******************************************************
plt.subplot(1,2,2)
plt.contourf(Y, X, ACT, 20, cmap = plt.cm.RdBu_r)
plt.title('Anomalia de Campo Total (nT)', fontsize = 12)
plt.xlabel('East (m)', fontsize = 10)
plt.ylabel('North (m)', fontsize = 10)
plt.colorbar()
#******************************************************
plt.show()
```

```python
# graficos
plt.close('all')
plt.figure(figsize=(17,6))
#******************************************************
plt.subplot(1,2,1)
plt.contourf(Y, X, CT2, 20, cmap = plt.cm.RdBu_r)
plt.title('Campo Total (nT)', fontsize = 12)
plt.xlabel('East (m)', fontsize = 10)
plt.ylabel('North (m)', fontsize = 10)
plt.colorbar()
#******************************************************
plt.subplot(1,2,2)
plt.contourf(Y, X, ACT2, 20, cmap = plt.cm.RdBu_r)
plt.title('Anomalia de Campo Total (nT)', fontsize = 12)
plt.xlabel('East (m)', fontsize = 10)
plt.ylabel('North (m)', fontsize = 10)
plt.colorbar()
#******************************************************
plt.show()
```
